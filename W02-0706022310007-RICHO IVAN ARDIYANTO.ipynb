**HANDLE MISSING VALUE FOR EACH COLUMN**

```python
# Mengisi missing value pada 'property_type' dengan mode (nilai yang paling sering muncul)
# Alasan: Kolom ini bersifat kategorikal, jadi lebih baik mengisi dengan nilai yang paling umum.
df['property_type'] = df['property_type'].fillna(df['property_type'].mode()[0])

# Periksa hasil pengisian
df['property_type'].isnull().sum()
```

```python
# Mengisi missing value pada 'bedrooms' dan 'bathrooms' dengan median
# Alasan: Median lebih stabil daripada mean jika ada outliers dalam data (seperti jumlah kamar yang sangat banyak).
df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())
df['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].median())

# Periksa hasil pengisian
df[['bedrooms', 'bathrooms']].isnull().sum()
```

```python
# Mengisi missing value pada 'land_size_m2' dan 'building_size_m2' dengan median
# Alasan: Untuk ukuran tanah dan bangunan, median lebih baik digunakan untuk menghindari pengaruh data yang ekstrem.
df['land_size_m2'] = df['land_size_m2'].fillna(df['land_size_m2'].median())
df['building_size_m2'] = df['building_size_m2'].fillna(df['building_size_m2'].median())

# Periksa hasil pengisian
df[['land_size_m2', 'building_size_m2']].isnull().sum()
```

```python
# Mengisi missing value pada 'certificate' dengan "Unknown"
# Alasan: Karena banyak data yang hilang dan kolom ini bersifat kategorikal, lebih baik diisi dengan "Unknown".
df['certificate'] = df['certificate'].fillna('Unknown')

# Periksa hasil pengisian
df['certificate'].isnull().sum()
```

```python
# Mengisi missing value pada 'building_age' dan 'year_built' dengan median
# Alasan: Median lebih tahan terhadap outliers dan lebih representatif untuk data numerik seperti usia bangunan.
df['building_age'] = df['building_age'].fillna(df['building_age'].median())
df['year_built'] = df['year_built'].fillna(df['year_built'].median())

# Periksa hasil pengisian
df[['building_age', 'year_built']].isnull().sum()
```

```python
# Mengisi missing value pada 'property_condition' dengan mode atau 'Unknown'
# Alasan: Karena kolom ini kategorikal, lebih baik diisi dengan nilai yang paling sering muncul atau dengan 'Unknown'.
df['property_condition'] = df['property_condition'].fillna(df['property_condition'].mode()[0])

# Periksa hasil pengisian
df['property_condition'].isnull().sum()
```

```python
# Mengisi missing value pada 'building_orientation' dengan 'Unknown'
# Alasan: Karena sebagian besar data hilang, lebih baik mengisinya dengan 'Unknown' daripada mengasumsikan nilai tertentu.
df['building_orientation'] = df['building_orientation'].fillna('Unknown')

# Periksa hasil pengisian
df['building_orientation'].isnull().sum()
```

```python
# Mengisi missing value pada 'furnishing' dengan 'Unknown'
# Alasan: Banyak data yang hilang, sehingga diisi dengan 'Unknown' untuk menjaga konsistensi.
df['furnishing'] = df['furnishing'].fillna('Unknown')

# Periksa hasil pengisian
df['furnishing'].isnull().sum()
```

```python
# Bersihkan kolom city (trim spasi) Alasan: Konsistensi label kota agar agregasi/Grouping tidak terpecah karena spasi.
df['city'] = df['city'].str.strip()
df['city'].unique()[:10]
```

```python
# Konversi electricity → numerik (VA/Watt) Alasan: Nilai seperti '4400 mah' tidak numerik. Kita ekstrak digitnya agar bisa dianalisis.
import re
df['electricity_va'] = (
    df['electricity']
    .astype(str)
    .str.extract(r'(\d+)')      # ambil angka
    .astype(float)
)
# Opsional: jika ingin drop kolom asli agar tidak membingungkan
# df = df.drop(columns=['electricity'])
df[['electricity', 'electricity_va']].head()
```

```python
# Validasi & perapihan year_built Alasan: Tahun di luar rentang wajar (mis. <1950 atau >2025) di-set NaN agar tidak bias.
import numpy as np

valid_min, valid_max = 1950, 2025
df['year_built_clean'] = df['year_built'].where(
    (df['year_built'] >= valid_min) & (df['year_built'] <= valid_max),
    np.nan
)
df[['year_built', 'year_built_clean']].head(10)
```

```python
# Sinkronisasi building_age ↔ year_built Alasan: Age seharusnya konsisten dengan 2025 - year_built. Prioritaskan tahun bila valid.
CURRENT_YEAR = 2025

# Hitung age dari year_built_clean jika tersedia
age_from_year = CURRENT_YEAR - df['year_built_clean']

# Jika building_age tidak logis (<0 atau >100), set NaN
df['building_age_clean'] = df['building_age'].where(
    (df['building_age'] >= 0) & (df['building_age'] <= 100),
    np.nan
)

# Isi building_age_clean dengan age_from_year jika age_from_year valid
df.loc[df['building_age_clean'].isna() & age_from_year.notna(), 'building_age_clean'] = age_from_year

df[['year_built', 'year_built_clean', 'building_age', 'building_age_clean']].head(10)
```

```python
# Sanitasi bedrooms & bathrooms (nilai ekstrem) Alasan: Nilai >10 kamar sangat jarang untuk rumah tapak; set NaN lalu imputasi median agar robust.
for col in ['bedrooms', 'bathrooms']:
    df[col] = df[col].where(df[col] <= 10, np.nan)
    df[col] = df[col].fillna(df[col].median())

df[['bedrooms', 'bathrooms']].describe()
```

```python
#  Sanitasi floorsAlasan: Isi lantai yang hilang dengan median (umumnya 1–2 lantai pada rumah).
df['floors'] = df['floors'].fillna(df['floors'].median())
df['floors'].value_counts(dropna=False)
```

```python
# Cap nilai ekstrem garages & carports Alasan: Nilai seperti 50 slot garasi sangat tidak realistis; cap di 6 (rumah besar).
df['garages']  = df['garages'].clip(upper=6)
df['carports'] = df['carports'].clip(upper=6)

df[['garages','carports']].describe()
```

```python
#  Tangani outlier land_size_m2 & building_size_m2 (IQR winsorization) Alasan: Winsorize via IQR menjaga informasi tanpa menghapus data, bagus untuk model robust.
def winsorize_iqr(s, k=1.5):
    q1, q3 = s.quantile(0.25), s.quantile(0.75)
    iqr = q3 - q1
    lower, upper = q1 - k*iqr, q3 + k*iqr
    return s.clip(lower=lower, upper=upper)

for col in ['land_size_m2', 'building_size_m2']:
    df[col] = winsorize_iqr(df[col].astype(float), k=1.5)

df[['land_size_m2', 'building_size_m2']].describe([.01,.05,.25,.5,.75,.95,.99])
```

```python
#  Perapihan kategori: certificate, furnishing, building_orientation Alasan: Pastikan tidak ada NaN; gunakan 'Unknown' agar model/EDA tetap bisa memproses kategori.
for col in ['certificate', 'furnishing', 'building_orientation', 'property_condition']:
    df[col] = df[col].fillna('Unknown')

df[['certificate','furnishing','building_orientation','property_condition']].head()
```

**Create new feature using binning**

```python
# Binning harga rumah (price_in_rp) ke dalam 3 kategori (Low, Medium, High) menggunakan qcut
df['price_bin'] = pd.qcut(df['price_in_rp'], q=3, labels=['Low', 'Medium', 'High'])

# Menampilkan kategori harga yang terbentuk
df['price_bin'].value_counts()
```

```python
# Binning jumlah lantai (floors) ke dalam kategori: 1 lantai, 2-3 lantai, lebih dari 3 lantai
df['floors_bin'] = pd.cut(df['floors'], bins=[0, 1, 3, df['floors'].max()],
                          labels=['1 Lantai', '2-3 Lantai', 'Lebih dari 3 Lantai'],
                          include_lowest=True)

# Menampilkan kategori jumlah lantai yang terbentuk
df['floors_bin'].value_counts()
```

```python
# Binning jumlah garasi (garages) ke dalam kategori: sedikit, sedang, banyak
df['garages_bin'] = pd.cut(df['garages'], bins=[0, 2, 4, df['garages'].max()],
                           labels=['Sedikit', 'Sedang', 'Banyak'],
                           include_lowest=True)

# Menampilkan kategori jumlah garasi yang terbentuk
df['garages_bin'].value_counts()
```

```python
# Binning luas tanah (land_size_m2) ke dalam 4 kategori: kecil, sedang, besar, sangat besar
df['land_bin'] = pd.cut(df['land_size_m2'],
                        bins=[0, 72, 150, 300, df['land_size_m2'].max()],
                        labels=['Kecil', 'Sedang', 'Besar', 'Sangat Besar'],
                        include_lowest=True)

# Menampilkan kategori luas tanah yang terbentuk
df['land_bin'].value_counts()
```

```python
# Binning luas bangunan (building_size_m2) ke dalam 4 kategori: kecil, sedang, besar, sangat besar
df['building_size_bin'] = pd.cut(df['building_size_m2'],
                                  bins=[0, 60, 100, 200, df['building_size_m2'].max()],
                                  labels=['Kecil', 'Sedang', 'Besar', 'Sangat Besar'],
                                  include_lowest=True)

# Menampilkan kategori luas bangunan yang terbentuk
df['building_size_bin'].value_counts()
```

```python
# Binning usia bangunan (building_age) menjadi kategori baru, sedang, tua
df['age_bin'] = pd.cut(df['building_age'],
                       bins=[0, 5, 10, 20, 50, df['building_age'].max()],
                       labels=['Baru', 'Sedang', 'Tua', 'Sangat Tua', 'Antik'],
                       include_lowest=True)

# Menampilkan kategori usia bangunan yang terbentuk
df['age_bin'].value_counts()
```

```python
# Binning furnitur (furnishing) ke dalam kategori Unfurnished, Furnished, Semi Furnished, Unknown
df['furnishing_bin'] = df['furnishing'].replace({'unfurnished': 'Unfurnished',
                                                 'semi furnished': 'Semi Furnished',
                                                 'furnished': 'Furnished',
                                                 'Unknown': 'Unknown'})

# Menampilkan kategori furnishing yang terbentuk
df['furnishing_bin'].value_counts()
```

```python
# Binning kondisi properti (property_condition) menjadi kategori kondisi properti
df['property_condition_bin'] = df['property_condition'].replace({'bagus': 'Bagus',
                                                                 'bagus sekali': 'Bagus Sekali',
                                                                 'baru': 'Baru',
                                                                 'sudah renovasi': 'Sudah Renovasi',
                                                                 'butuh renovasi': 'Butuh Renovasi',
                                                                 'semi furnished': 'Semi Furnished',
                                                                 'unfurnished': 'Unfurnished'})

# Menampilkan kategori kondisi properti yang terbentuk
df['property_condition_bin'].value_counts()
```

```python
# Binning kolom 'certificate' menjadi kategori yang lebih umum
df['certificate_bin'] = df['certificate'].replace({
    'shm - sertifikat hak milik': 'Hak Milik (SHM)',
    'hgb - hak guna bangunan': 'Hak Guna Bangunan (HGB)',
    'hp - hak pakai': 'Hak Pakai (HP)',
    'lainnya (ppjb,girik,adat,dll)': 'Lainnya',
    'Unknown': 'Unknown'
})

# Menampilkan hasil kategori sertifikat yang terbentuk
df['certificate_bin'].value_counts()
```
